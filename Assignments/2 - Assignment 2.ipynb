### 21. Determine and comment on the day of the week that movies are released. What is notable about release days and months throughout the year? (Use pd.pivot_table)

```python
release_dates_copy = release_dates.copy()
release_dates_copy['date'] = pd.to_datetime(release_dates_copy['date'])
release_dates_copy['weekday'] = release_dates_copy['date'].dt.day_name()
release_dates_copy['month'] = release_dates_copy['date'].dt.month_name()

pivot_weekday = pd.pivot_table(release_dates_copy, values='title', index='weekday', aggfunc='count').reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']) # sort rows manually

pivot_weekday
```

<span style="color:green">
<b>

The pivot table shows that Friday is the most popular movie release day, followed by Thursday. This is common practice in many countries, with the goal of maximizing box office earnings over the weekend. The least popular movie release days are Sunday and Tuesday.

</b>
</span>

```python
pivot_month = pd.pivot_table(release_dates_copy, values='title', index='month', aggfunc='count').reindex([
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'])  # sort rows manually

pivot_month
```

<span style="color:green">
<b>

Surprisingly, March is shown to be the most popular movie release month. I'm not entirely sure why this is the case, but perhaps some Hollywood films are initially released in the U.S. around January and February, then in many other countries in March. October is the second most popular movie release month, followed by a few winter months (January, December, and November). This makes more sense because there are several major holidays (e.g., Halloween, Thanksgiving, and Christmas) during these months when a lot of movies are usually released. Moreover, many films contending for awards are typically released around this time each year.

</b>
</span>
